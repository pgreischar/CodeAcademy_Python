Multiple Line Plots in Single axis

```python
# Days of the week:
days = [0, 1, 2, 3, 4, 5, 6]
# Your Money:
money_spent = [17, 12, 12, 10, 14, 16, 10]
# Your Friend's Money:
money_spent_2 = [11, 4, 5, 15, 22, 21, 22]
# Plot your money:
plt.plot(days, money_spent)
# Plot your friend's money: (notice the different line style formatting)
plt.plot(days, money_spent_2,linestyle=':')
# Above in linestyle you could also use a -- to be a dashed line.
# Display the result:
plt.show()
```
